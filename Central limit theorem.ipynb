```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# uniform distribution

num_iter = 10 ** 5
n = 10
mu = 0.5
sigma = np.sqrt(1 / 12)

result = np.empty(shape = num_iter)
for i in range(0, num_iter):
    u = np.random.uniform(0, 1, size = 20)
    mean_u = np.mean(u)
    z = (mean_u - mu) / (sigma / np.sqrt(n))
    result[i] = z
```

```python
plt.hist(result,
        bins = 50,
        density = True)
plt.title('Histogram of sample mean from U(0,1) with n = {}'. format(n))
plt.show()
```

```python
# exponential distribution

num_iter = 10 ** 5
n = 10
theta = 5
sigma = theta ** 2

result = np.empty(shape = num_iter)
for i in range(0, num_iter):
    u = np.random.exponential(scale = theta, size = n)
    mean_u = np.mean(u)
    z = (mean_u - theta) / (sigma / np.sqrt(n))
    result[i] = z
```

```python
plt.hist(result,
        bins = 50,
        density = True)
plt.title('Histogram of sample mean from EXP({}) with n = {}'. format(theta, n))
plt.show()
```
